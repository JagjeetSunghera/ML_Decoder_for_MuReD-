```python
import cv2

def retinal_crop(filename,destfilename):
    # Load image, convert to grayscale, and find edges
    image = cv2.imread(filename)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU + cv2.THRESH_BINARY)[1]

    # Find contour and sort by contour area
    cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = cnts[0] if len(cnts) == 2 else cnts[1]
    cnts = sorted(cnts, key=cv2.contourArea, reverse=True)
    for c in cnts:
        # draw filled circles in white on black background as masks
        x,y,w,h = cv2.boundingRect(c)
#         print(x,y,w,h)
        result=image.copy()
#         print(image.shape)
        mask1 = np.zeros_like(result)
        mask = cv2.circle(mask1, (int(x+(w/2)),int(y+(h/2))), int(max(h,w)/2), (255,255,255), -1)
        result = cv2.bitwise_and(result,mask)
#         result=result[y:y+max(h,w),x:x+max(h,w)]
        break
    cv2.imwrite(destfilename,result)
```

```python
import cv2

def centercrop(filename,destfilename):
    # Load image, convert to grayscale, and find edges
    image = cv2.imread(filename)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU + cv2.THRESH_BINARY)[1]

    # Find contour and sort by contour area
    cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = cnts[0] if len(cnts) == 2 else cnts[1]
    cnts = sorted(cnts, key=cv2.contourArea, reverse=True)
    for c in cnts:
        # draw filled circles in white on black background as masks
        x,y,w,h = cv2.boundingRect(c)
        result=image.copy()
        result=result[y:y+h,x:x+w]
        break
    cv2.imwrite(destfilename,result)
    
    
    
def padding(filename,destfilename):
    # Load image, convert to grayscale, and find edges
    img = cv2.imread(filename)
    old_image_height, old_image_width, channels = img.shape
    # create new image of desired size and color (blue) for padding
    new_image_width = max(img.shape)
    new_image_height = max(img.shape)
    color = (0,0,0)
    result = np.full((new_image_height,new_image_width, channels), color, dtype=np.uint8)
    # compute center offset
    x_center = (new_image_width - old_image_width) // 2
    y_center = (new_image_height - old_image_height) // 2
    # copy img image into center of result image
    result[y_center:y_center+old_image_height, 
           x_center:x_center+old_image_width] = img
    cv2.imwrite(destfilename,result)
    
    
def Resizecrop(filename,destination):
    augment=transforms.Compose([transforms.Resize(512,antialias=True),transforms.CenterCrop(512)])
    image=Image.open(filename)
    image=augment(image)
    image.save(destination)
```
